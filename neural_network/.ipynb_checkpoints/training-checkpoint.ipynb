# Prepare the data

```python
import ROOT

# def filter_events(df):
#     """
#     Reduce initial dataset to only events which shall be used for training
#     """
#     return df.Filter("","Dimuon bigger than 3")
 
def define_variables(df):
    """
    Define the variables which shall be used for training
    """
    return df.Define("DimuonMass", "Dimuon_Mass")\
             .Define("Bz_Eta", "B_Eta")\
             .Define("Bz_Mass", "B_Mass")\
             .Define("KstarMass", "Kstar_Mass")\
             .Define("MuonpT1", "Muon1_pT")\
             .Define("MuonpT2", "Muon2_pT")\
             .Define("trackpT1", "track1_pT")\
             .Define("trackpT2", "track2_pT")

             #eta B, no eta particelle
             #no phi
             #Kstar mass si e tutti pT
             #escludi non riempite
                

variables = ["Dimuon_Mass", "Bz_Eta", "Bz_Mass"
             ,"Kstar_Mass","Muon1_pT","Muon2_pT", "track1_pT", "track2_pT"]
 
if __name__ == "__main__":
#     for filename, label in [["Data-BParking_v0.root", "data"], ["BdToJpsiKstar_BMuonFilter_v0.root", "MC"]]:
#         print(">>> Extract the training and testing events for {} from the {} dataset.".format(
#             label, filename))

        #data 
        print("Preparing data...")
        filepath = "new_data.root"
        label = "background"
        # Load dataset, filter the required events and define the training variables
        df = ROOT.RDataFrame("ntupleTree", filepath)
#         df = fiblter_events(df)
        df = define_variables(df)
 
        # Book cutflow report
        report = df.Report()
 
        # Save the filtered dataset 
        columns = ROOT.std.vector["string"](variables)
        df.Snapshot("demo/ntupleTree", "train_" + label + ".root", columns)
 
        # Print cutflow report
        report.Print()
        print ("File " +filepath+" created!")
        
        #mc 
        print("Preparing Montecarlo...")
        filepath = "new_MC.root"
        label = "signal"
        # Load dataset, filter the required events and define the training variables
        df = ROOT.RDataFrame("ntupleTree", filepath)
#         df = filter_events(df)
        df = define_variables(df)
 
        # Book cutflow report
        report = df.Report()
 
        # Save the filtered dataset 
        columns = ROOT.std.vector["string"](variables)
        df.Snapshot("demo/ntupleTree", "train_" + label + ".root", columns)
 
        # Print cutflow report
        report.Print()
        print ("File " +filepath+" created!")
```

## Train the data

```python
data_sig = ROOT.RDataFrame("demo/ntupleTree", "train_signal.root").AsNumpy()
```

```python
def load_data(signal_filename, background_filename):
    # Read data from ROOT files
    data_sig = ROOT.RDataFrame("demo/ntupleTree", signal_filename).AsNumpy()
    data_bkg = ROOT.RDataFrame("demo/ntupleTree", background_filename).AsNumpy()

    # Convert inputs to format readable by machine learning tools
    x_sig = np.vstack([data_sig[var] for var in variables]).T
    x_bkg = np.vstack([data_bkg[var] for var in variables]).T
    x = np.vstack([x_sig, x_bkg])

    # Create labels
    num_sig = x_sig.shape[0]
    num_bkg = x_bkg.shape[0]
    y = np.hstack([np.ones(num_sig), np.zeros(num_bkg)])

    # Compute weights balancing both classes
    num_all = num_sig + num_bkg
    w = np.hstack([np.ones(num_sig) * num_all / num_sig, np.ones(num_bkg) * num_all / num_bkg])

    return x, y, w

if __name__ == "__main__":
    # Load data
    x, y, w = load_data("train_signal.root", "train_background.root")

    # Fit xgboost model
    from xgboost import XGBClassifier
    bdt = XGBClassifier(max_depth=3, n_estimators=500, use_label_encoder=False)
    bdt.fit(x, y, w)

    # Save model in TMVA format
    ROOT.TMVA.Experimental.SaveXGBoost(bdt, "myBDT", "tmva.root", num_inputs=len(variables))
```

## Test 

```python
# Load data
x, y_true, w = load_data("OG_data/BdToJpsiKstar_BMuonFilter_v0.root", "OG_data/Data-BParking_v0.root")

# Load trained model
bdt = ROOT.TMVA.Experimental.RBDT[""]("myBDT", "tmva.root")

# Make prediction
y_pred = bdt.Compute(x)

# Compute ROC using sklearn
from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(y_true, y_pred, sample_weight=w)
score = auc(fpr, tpr)

# Plot ROC
c = ROOT.TCanvas("roc", "", 600, 600)
g = ROOT.TGraph(len(fpr), fpr, tpr)
g.SetTitle("AUC = {:.2f}".format(score))
g.SetLineWidth(3)
g.SetLineColor(ROOT.kRed)
g.Draw("AC")
g.GetXaxis().SetRangeUser(0, 1)
g.GetYaxis().SetRangeUser(0, 1)
g.GetXaxis().SetTitle("False-positive rate")
g.GetYaxis().SetTitle("True-positive rate")
c.Draw()
```
